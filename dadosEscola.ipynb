```python
#Pegando os dados das colunas referente as escolas cadastradas 
colunasDadosEscola = ['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC',
       'CO_UF_ESC', 'SG_UF_ESC', 'TP_DEPENDENCIA_ADM_ESC',
       'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC']
```

```python
#Filtrando as colunas que irei utilizar

DadosParticipantes = df.filter(items=colunasDadosEscola)
```

```python
dropDuplicados = DadosParticipantes.drop_duplicates()
```

```python
# Removendo registros onde toda a linha está como NaN

dadosLimpos = dropDuplicados.dropna(how='all')

dadosLimpos.head()
```

```python
dadoslimpos['CO_MUNICIPIO_ESC'] = dadoslimpos['CO_MUNICIPIO_ESC'].fillna(0).astype(int)
dadoslimpos['CO_UF_ESC'] = dadoslimpos['CO_UF_ESC'].fillna(0).astype(int) 
dadoslimpos['TP_DEPENDENCIA_ADM_ESC'] = dadoslimpos['TP_DEPENDENCIA_ADM_ESC'].fillna(0).astype(int)
dadoslimpos['TP_LOCALIZACAO_ESC'] = dadoslimpos['TP_LOCALIZACAO_ESC'].fillna(0).astype(int)
dadoslimpos['TP_SIT_FUNC_ESC'] = dadoslimpos['TP_SIT_FUNC_ESC'].fillna(0).astype(int)
```

```python
dadosLimpos.dtypes
```

```python
#Tratando dados
def editar_tp_dependencia_adm_esc(tp_dependencia_adm_esc):
    if tp_dependencia_adm_esc == 1:
        return 'Federal'
    elif tp_dependencia_adm_esc == 2:
        return 'Estadual'
    elif tp_dependencia_adm_esc == 3:
        return 'Municipal'
    elif tp_dependencia_adm_esc == 4:
        return 'Privada'
    else:
        return 'Não informado'
    
def editar_tp_localizacao_esc(tp_localizacao_esc):
    if tp_localizacao_esc == 1:
        return 'Urbana'
    elif tp_localizacao_esc == 2:
        return 'Rural'
    else:
        return 'Não informado'

def editar_tp_sit_func_esc(tp_sit_func_esc):
    if tp_sit_func_esc == 1:
        return 'Em atividade'
    elif tp_sit_func_esc == 2:
        return 'Paralisada'
    elif tp_sit_func_esc == 3:
        return 'Extinta'
    elif tp_sit_func_esc == 4:
        return 'Escola extinta em anos anteriores'
    else:
        return 'Não informado'
```

```python
def tratar_dados(dadosLimpos):
    dadosLimpos = dadosLimpos.copy()
    dadosLimpos['TP_DEPENDENCIA_ADM_ESC'] = dadosLimpos['TP_DEPENDENCIA_ADM_ESC'].apply(editar_tp_dependencia_adm_esc)
    dadosLimpos['TP_LOCALIZACAO_ESC'] = dadosLimpos['TP_LOCALIZACAO_ESC'].apply(editar_tp_localizacao_esc)
    dadosLimpos['TP_SIT_FUNC_ESC'] = dadosLimpos['TP_SIT_FUNC_ESC'].apply(editar_tp_sit_func_esc)

    return dadosLimpos
```

```python
# aplica a função de tratamento dos dados
dadosLimpos = tratar_dados(dadosLimpos)
```

```python
import pyodbc
import pandas as pd
from app.db_config import database_config

# Criar a string de conexão diretamente para `pyodbc`
conn_str = (
    f"DRIVER=ODBC Driver 17 for SQL Server;"
    f"SERVER={database_config['host']},{database_config['port']};"
    f"DATABASE={database_config['database']};"
    f"UID={database_config['username']};"
    f"PWD={database_config['password']};"
    "TrustServerCertificate=yes"
)

# Criar a conexão diretamente com `pyodbc`
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

#garantir que as colunas do tipo "Object" sejam convertidas para string
dadosLimpos = dadosLimpos.convert_dtypes()

for col in dadosLimpos.select_dtypes(include=['object']).columns:
    dadosLimpos[col] = dadosLimpos[col].astype(str)

print(dadosLimpos.dtypes)

# Remover duplicatas no novo dataset
dadosLimpos = dadosLimpos.drop_duplicates()


try:
    #criando a query de inserção dinâmica
    insert_query = """
        INSERT INTO [dbo].[DadosEscola]
           (CO_MUNICIPIO_ESC
           ,NO_MUNICIPIO_ESC
           ,CO_UF_ESC
           ,SG_UF_ESC
           ,TP_DEPENDENCIA_ADM_ESC
           ,TP_LOCALIZACAO_ESC
           ,TP_SIT_FUNC_ESC)
        VALUES (?, ?, ?, ?, ?, ?, ?)
    """

    # converte o dataframe para Lista de Tuplas 
    records = [tuple(x) for x in dadosLimpos.to_numpy()] #converte corretamente para lista de tuplas

    #cursor para inserrir os dados em lote
    cursor.executemany(insert_query, records)

    #confirma transação
    conn.commit()

    print("✅ Dados inseridos com sucesso!")

except Exception as e:
    print("❌ Erro ao inserir os dados:", e)
```
