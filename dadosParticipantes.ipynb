```python
colunasDadosParticipantes = ['NU_INSCRICAO', 'NU_ANO', 'TP_FAIXA_ETARIA', 'TP_SEXO',
       'TP_ESTADO_CIVIL', 'TP_COR_RACA', 'TP_NACIONALIDADE',
       'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO',
       'IN_TREINEIRO']
```

```python
colunasDadosParticipantes = df.filter(items=colunasDadosParticipantes)
```

```python
  
def editar_faixaEtaria(faixa_etaria):
    if faixa_etaria == 1:
        return "Menor de 17 anos"
    elif faixa_etaria == 2:
        return "17 anos"
    elif faixa_etaria == 3:
        return "18 anos"
    elif faixa_etaria == 4:
        return "19 anos"
    elif faixa_etaria == 5:
        return "20 anos"
    elif faixa_etaria == 6:
        return "21 anos"
    elif faixa_etaria == 7:
        return "22 anos"
    elif faixa_etaria == 8:
        return "23 anos"
    elif faixa_etaria == 9:
        return "24 anos"
    elif faixa_etaria == 10:
        return "25 anos"
    elif faixa_etaria == 11:
        return "Entre 26 e 30 anos"
    elif faixa_etaria == 12:
        return "Entre 31 e 35 anos"
    elif faixa_etaria == 13:
        return "Entre 36 e 40 anos"
    elif faixa_etaria == 14:
        return "Entre 41 e 45 anos"
    elif faixa_etaria == 15:
        return "Entre 46 e 50 anos"
    elif faixa_etaria == 16:
        return "Entre 51 e 55 anos"
    elif faixa_etaria == 17:
        return "Entre 56 e 60 anos"
    elif faixa_etaria == 18:
        return "Entre 61 e 65 anos"
    elif faixa_etaria == 19:
        return "Entre 66 e 70 anos"
    elif faixa_etaria == 20:
        return "Maior de 70 anos"
    else:
        return "Não Informado"

def editar_sexo(sexo):
    if sexo == 'M':
        return "Masculino"
    else:
        return "Feminino"

def editar_estado_civil(estado_civil):
    if estado_civil == 0:
        return "Não informado"
    elif estado_civil == 1:
        return "Solteiro(a)"
    elif estado_civil == 2:
        return "Casado(a)/Mora com companheiro(a)"
    elif estado_civil == 3:
        return "Divorciado(a)/Desquitado(a)/Separado(a)"
    else:
        return "Viúvo(a)"

def editar_raca_cor(raca_cor):
    if raca_cor == 0:
        return "Não declarado"
    elif raca_cor == 1:
        return "Branca"
    elif raca_cor == 2:
        return "Preta"
    elif raca_cor == 3:
        return "Parda"
    elif raca_cor == 4:
        return "Amarela"
    elif raca_cor == 5:
        return "Indígena "
    else:
        return "Não dispõe da informação"
    
def editar_nacionalidade(nacionalidade):
    if nacionalidade == 0:
        return "Não informado"
    elif nacionalidade == 0:
        return "Brasileiro(a)"
    elif nacionalidade == 0:
        return "Brasileiro(a) Naturalizado(a)"
    elif nacionalidade == 0:
        return "Estrangeiro(a)"
    elif nacionalidade == 0:
        return "Brasileiro(a) Nato(a), nascido(a) no exterior"
    else:
        return "Não informado"
    
def editar_st_conclusao(st_conclusao):
    if st_conclusao == 1:
        return "Já concluí o Ensino Médio"
    elif st_conclusao == 2:
        return "Estou cursando e concluirei o Ensino Médio em 2023"
    elif st_conclusao == 3:
        return "Estou cursando e concluirei o Ensino Médio após 2023"
    else:
        return "Não concluí e não estou cursando o Ensino Médio"


def editar_ano_conclusao(st_conclusao):
    if st_conclusao == 0:
        return "Não informado"
    elif st_conclusao == 1:
        return "2022"
    elif st_conclusao == 2:
        return "2021"
    elif st_conclusao == 3:
        return "2020"
    elif st_conclusao == 4:
        return "2019"
    elif st_conclusao == 5:
        return "2018"
    elif st_conclusao == 6:
        return "2017"
    elif st_conclusao == 7:
        return "2016"
    elif st_conclusao == 8:
        return "2015"
    elif st_conclusao == 9:
        return "2014"
    elif st_conclusao == 10:
        return "2013"
    elif st_conclusao == 11:
        return "2012"
    elif st_conclusao == 12:
        return "2011"
    elif st_conclusao == 13:
        return "2010"
    elif st_conclusao == 14:
        return "2009"
    elif st_conclusao == 15:
        return "2008"
    elif st_conclusao == 16:
        return "2007"
    else:
        return "Antes de 2007"

def editar_esino_medio(ensino_medio):
    if ensino_medio == 1:
        return "Não Respondeu"
    elif ensino_medio == 2:
        return "Pública"
    else:
        return "Privada"

def editar_ensino(ensino):
    if ensino == 1:
        return "Ensino Regular"
    elif ensino == 2:
        return "Educação Especial - Modalidade Substitutiva"
    else:
        return "Não preenchido" 

def editar_treineiro(treineiro):
    if treineiro == 1:
        return "Sim"
    else:
        return "Não"
```

```python
def tratar_dados(colunasDadosParticipantes):
    colunasDadosParticipantes['TP_FAIXA_ETARIA'] = colunasDadosParticipantes['TP_FAIXA_ETARIA'].apply(editar_faixaEtaria)
    colunasDadosParticipantes['TP_SEXO'] = colunasDadosParticipantes['TP_SEXO'].apply(editar_sexo)
    colunasDadosParticipantes['TP_ESTADO_CIVIL'] = colunasDadosParticipantes['TP_ESTADO_CIVIL'].apply(editar_estado_civil)
    colunasDadosParticipantes['TP_COR_RACA'] = colunasDadosParticipantes['TP_COR_RACA'].apply(editar_raca_cor)
    colunasDadosParticipantes['TP_NACIONALIDADE'] = colunasDadosParticipantes['TP_NACIONALIDADE'].apply(editar_nacionalidade)
    colunasDadosParticipantes['TP_ST_CONCLUSAO'] = colunasDadosParticipantes['TP_ST_CONCLUSAO'].apply(editar_st_conclusao)
    colunasDadosParticipantes['TP_ANO_CONCLUIU'] = colunasDadosParticipantes['TP_ANO_CONCLUIU'].apply(editar_ano_conclusao)
    colunasDadosParticipantes['TP_ESCOLA'] = colunasDadosParticipantes['TP_ESCOLA'].apply(editar_esino_medio)
    colunasDadosParticipantes['TP_ENSINO'] = colunasDadosParticipantes['TP_ENSINO'].apply(editar_ensino)
    colunasDadosParticipantes['IN_TREINEIRO'] = colunasDadosParticipantes['IN_TREINEIRO'].apply(editar_treineiro)
    
    return colunasDadosParticipantes
```

```python
# aplica a função de tratamento dos dados
dadosParticipantes = tratar_dados(colunasDadosParticipantes)
```
