```python
#tratamento de dados seguindo o dicionario disponibilizado

def editar_tp_presenca_cn(tp_presenca_cn):
    if tp_presenca_cn == 0:
        return 'Faltou à prova'
    elif tp_presenca_cn == 1:
        return 'Presente na prova'
    elif tp_presenca_cn == 2:
        return 'Eliminado na prova'
    else:
        return 'Faltou à prova'

def editar_tp_presenca_ch(tp_presenca_ch):
    if tp_presenca_ch == 0:
        return 'Faltou à prova'
    elif tp_presenca_ch == 1:
        return 'Presente na prova'
    elif tp_presenca_ch == 2:
        return 'Eliminado na prova'
    else:
        return 'Faltou à prova'
    
def editar_tp_presenca_lc(tp_presenca_lc):
    if tp_presenca_lc == 0:
        return 'Faltou à prova'
    elif tp_presenca_lc == 1:
        return 'Presente na prova'
    elif tp_presenca_lc == 2:
        return 'Eliminado na prova'
    else:
        return 'Faltou à prova'

def editar_tp_presenca_mt(tp_presenca_mt):
    if tp_presenca_mt == 0:
        return 'Faltou à prova'
    elif tp_presenca_mt == 1:
        return 'Presente na prova'
    elif tp_presenca_mt == 2:
        return 'Eliminado na prova'
    else:
        return 'Faltou à prova'
    
def editar_co_prova_cn(co_prova_cn):
    if co_prova_cn == 1221:
        return 'Azul'
    elif co_prova_cn == 1222:
        return 'Amarela'
    elif co_prova_cn == 1223:
        return 'Rosa'
    elif co_prova_cn == 1224:
        return 'Cinza'
    elif co_prova_cn == 1225:
        return 'Rosa - Ampliada'
    elif co_prova_cn == 1226:
        return 'Rosa - Superampliada'
    elif co_prova_cn == 1227:
        return 'Laranja - Braile'
    elif co_prova_cn == 1228:
        return 'Laranja - Adaptada Ledor'
    elif co_prova_cn == 1229:
        return 'Verde - Videoprova - Libras'
    elif co_prova_cn == 1301:
        return 'Azul (Reaplicação)'
    elif co_prova_cn == 1302:
        return 'Amarela (Reaplicação)'
    elif co_prova_cn == 1303:
        return 'Cinza (Reaplicação)'
    elif co_prova_cn == 1304:
        return 'Rosa (Reaplicação)'
    else:
        return 'Faltou à prova'
    
def editar_co_prova_ch(co_prova_ch):
    if co_prova_ch == 1191:
        return 'Azul'
    elif co_prova_ch == 1192:
        return 'Amarela'
    elif co_prova_ch == 1193:
        return 'Rosa'
    elif co_prova_ch == 1194:
        return 'Cinza'
    elif co_prova_ch == 1195:
        return 'Rosa - Ampliada'
    elif co_prova_ch == 1196:
        return 'Rosa - Superampliada'
    elif co_prova_ch == 1197:
        return 'Laranja - Braile'
    elif co_prova_ch == 1198:
        return 'Laranja - Adaptada Ledor'
    elif co_prova_ch == 1199:
        return 'Verde - Videoprova - Libras'
    elif co_prova_ch == 1271:
        return 'Azul (Reaplicação)'
    elif co_prova_ch == 1272:
        return 'Amarela (Reaplicação)'
    elif co_prova_ch == 1273:
        return 'Cinza (Reaplicação)'
    elif co_prova_ch == 1274:
        return 'Rosa (Reaplicação)'
    else:
        return 'Faltou à prova'
    
def editar_co_prova_lc(co_prova_lc):
    if co_prova_lc == 1201:
        return 'Azul'
    elif co_prova_lc == 1202:
        return 'Amarela'
    elif co_prova_lc == 1203:
        return 'Rosa'
    elif co_prova_lc == 1204:
        return 'Cinza'
    elif co_prova_lc == 1205:
        return 'Rosa - Ampliada'
    elif co_prova_lc == 1206:
        return 'Rosa - Superampliada'
    elif co_prova_lc == 1207:
        return 'Laranja - Braile'
    elif co_prova_lc == 1208:
        return 'Laranja - Adaptada Ledor'
    elif co_prova_lc == 1209:
        return 'Verde - Videoprova - Libras'
    elif co_prova_lc == 1281:
        return 'Azul (Reaplicação)'
    elif co_prova_lc == 1282:
        return 'Amarela (Reaplicação)'
    elif co_prova_lc == 1283:
        return 'Cinza (Reaplicação)'
    elif co_prova_lc == 1284:
        return 'Rosa (Reaplicação)'
    else:
        return 'Faltou à prova'
    
def editar_co_prova_mt(co_prova_mt):
    if co_prova_mt == 1211:
        return 'Azul'
    elif co_prova_mt == 1212:
        return 'Amarela'
    elif co_prova_mt == 1213:
        return 'Rosa'
    elif co_prova_mt == 121:
        return 'Cinza'
    elif co_prova_mt == 1215:
        return 'Rosa - Ampliada'
    elif co_prova_mt == 1216:
        return 'Rosa - Superampliada'
    elif co_prova_mt == 1217:
        return 'Laranja - Braile'
    elif co_prova_mt == 1218:
        return 'Laranja - Adaptada Ledor'
    elif co_prova_mt == 1219:
        return 'Verde - Videoprova - Libras'
    elif co_prova_mt == 1291:
        return 'Azul (Reaplicação)'
    elif co_prova_mt == 1292:
        return 'Amarela (Reaplicação)'
    elif co_prova_mt == 1293:
        return 'Cinza (Reaplicação)'
    elif co_prova_mt == 1294:
        return 'Rosa (Reaplicação)'
    else:
        return 'Faltou à prova'
    
def editar_tp_lingua(row):
    tp_lingua = row['TP_LINGUA']
    tp_presenca_cn = row['TP_PRESENCA_CN']
    tp_presenca_ch = row['TP_PRESENCA_CH']
    tp_presenca_lc = row['TP_PRESENCA_LC']
    tp_presenca_mt = row['TP_PRESENCA_MT']
    
    if tp_presenca_cn == 1 and tp_presenca_ch == 1 and tp_presenca_lc == 1 and tp_presenca_mt == 1 and tp_lingua == 0:
        return 'Inglês'
    elif tp_presenca_cn == 1 and tp_presenca_ch == 1 and tp_presenca_lc == 1 and tp_presenca_mt == 1 and tp_lingua == 1:
        return 'Espanhol'
    else:
        return 'Faltou à prova'
    
def editar_tx_respostas(row):
    tx_respostas_cn = row['TX_RESPOSTAS_CN']
    tx_respostas_ch = row['TX_RESPOSTAS_CH']
    tx_respostas_lc = row['TX_RESPOSTAS_LC']
    tx_respostas_mt = row['TX_RESPOSTAS_MT']
    
    # Verifica se os valores são nulos ou vazios e retorna 'Faltou à prova' se necessário
    tx_respostas_cn = 'Faltou à prova' if (pd.isnull(tx_respostas_cn) or tx_respostas_cn == '') else tx_respostas_cn
    tx_respostas_ch = 'Faltou à prova' if (pd.isnull(tx_respostas_ch) or tx_respostas_ch == '') else tx_respostas_ch
    tx_respostas_lc = 'Faltou à prova' if (pd.isnull(tx_respostas_lc) or tx_respostas_lc == '') else tx_respostas_lc
    tx_respostas_mt = 'Faltou à prova' if (pd.isnull(tx_respostas_mt) or tx_respostas_mt == '') else tx_respostas_mt
    
    # Retorna uma combinação dos resultados ou apenas o valor de interesse
    return tx_respostas_cn, tx_respostas_ch, tx_respostas_lc, tx_respostas_mt

def editar_nota_prova(row):
    nu_nota_cn = row['NU_NOTA_CN']
    nu_nota_ch = row['NU_NOTA_CH']
    nu_nota_lc = row['NU_NOTA_LC']
    nu_nota_mt = row['NU_NOTA_MT']

    nu_nota_cn = 0 if (pd.isnull(nu_nota_cn) or nu_nota_cn == '') else nu_nota_cn
    nu_nota_ch = 0 if (pd.isnull(nu_nota_ch) or nu_nota_ch == '') else nu_nota_ch
    nu_nota_lc = 0 if (pd.isnull(nu_nota_lc) or nu_nota_lc == '') else nu_nota_lc
    nu_nota_mt = 0 if (pd.isnull(nu_nota_mt) or nu_nota_mt == '') else nu_nota_mt

    # Retorna uma combinação dos resultados ou apenas o valor de interesse
    return nu_nota_cn, nu_nota_ch, nu_nota_lc, nu_nota_mt

def editar_tx_gabarito(row):
    tx_gabarito_cn = row['TX_GABARITO_CN']
    tx_gabarito_ch = row['TX_GABARITO_CH']
    tx_gabarito_lc = row['TX_GABARITO_LC']
    tx_gabarito_mt = row['TX_GABARITO_MT']

    tx_gabarito_cn = 'Faltou à prova' if (pd.isnull(tx_gabarito_cn) or tx_gabarito_cn == '') else tx_gabarito_cn
    tx_gabarito_ch = 'Faltou à prova' if (pd.isnull(tx_gabarito_ch) or tx_gabarito_ch == '') else tx_gabarito_ch
    tx_gabarito_lc = 'Faltou à prova' if (pd.isnull(tx_gabarito_lc) or tx_gabarito_lc == '') else tx_gabarito_lc
    tx_gabarito_mt = 'Faltou à prova' if (pd.isnull(tx_gabarito_mt) or tx_gabarito_mt == '') else tx_gabarito_mt

    # Retorna uma combinação dos resultados ou apenas o valor de interesse
    return tx_gabarito_cn, tx_gabarito_ch, tx_gabarito_lc, tx_gabarito_mt
```

```python
def tratardados(dadosLimpos):
    dadosLimpos['TP_PRESENCA_CN'] = dadosLimpos['TP_PRESENCA_CN'].apply(editar_tp_presenca_cn)
    dadosLimpos['TP_PRESENCA_CH'] = dadosLimpos['TP_PRESENCA_CH'].apply(editar_tp_presenca_ch)
    dadosLimpos['TP_PRESENCA_LC'] = dadosLimpos['TP_PRESENCA_LC'].apply(editar_tp_presenca_lc)
    dadosLimpos['TP_PRESENCA_MT'] = dadosLimpos['TP_PRESENCA_MT'].apply(editar_tp_presenca_mt)
    dadosLimpos['CO_PROVA_CN'] = dadosLimpos['CO_PROVA_CN'].apply(editar_co_prova_cn)
    dadosLimpos['CO_PROVA_CH'] = dadosLimpos['CO_PROVA_CH'].apply(editar_co_prova_ch)
    dadosLimpos['CO_PROVA_LC'] = dadosLimpos['CO_PROVA_LC'].apply(editar_co_prova_lc)
    dadosLimpos['CO_PROVA_MT'] = dadosLimpos['CO_PROVA_MT'].apply(editar_co_prova_mt)
    dadosLimpos['TP_LINGUA'] = dadosLimpos.apply(editar_tp_lingua, axis=1)
    dadosProva_objetiva[['TX_RESPOSTAS_CN', 'TX_RESPOSTAS_CH', 'TX_RESPOSTAS_LC', 'TX_RESPOSTAS_MT']] = dadosProva_objetiva.apply(lambda row: pd.Series(editar_tx_respostas(row)), axis=1)
    dadosProva_objetiva[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT']] = dadosProva_objetiva.apply(lambda row: pd.Series(editar_nota_prova(row)), axis=1)
    dadosProva_objetiva[['TX_GABARITO_CN', 'TX_GABARITO_CH', 'TX_GABARITO_LC', 'TX_GABARITO_MT']] = dadosProva_objetiva.apply(lambda row: pd.Series(editar_tx_gabarito(row)), axis=1)
    
    return dadosLimpos

```

```python
# Aplica a função de tratamento dos dados
dadosProvasObjLimpos = tratardados(dadosProva_objetiva)
```

```python
#Descobrindo o tamanho maximo do campo para criar a tabela de forma correta.
dadosProvasObjLimpos.dtypes
```

```python
tabela = 'DadosProvaObjetiva'

tipo_dado_map = {
    'int64': 'BIGINT',
    'float64': 'FLOAT',
    'object': 'VARCHAR({})',
    'bool': 'BIT',
    'datetime64[ns]': 'DATETIME'
}

def calcular_max_varchar_colunas(dadosProvasObjLimpos, coluna):
    
    #retornar o comprimento máximo da coluna
    if dadosProvasObjLimpos[coluna].dtype == 'object':
        return dadosProvasObjLimpos[coluna].apply(lambda x: len(str(x))).max()
    return None


def gerar_sql_create_table(dadosProvasObjLimpos, tabela):
    colunas = []
    
    for coluna, tipo in dadosProvasObjLimpos.dtypes.items():
        if str(tipo) in tipo_dado_map:
            tipo_sql = tipo_dado_map[str(tipo)]
            if 'VARCHAR' in tipo_sql:
                max_length = calcular_max_varchar_colunas(dadosProvasObjLimpos, coluna)
                if max_length is None:
                    max_length = 255  # Valor padrão caso não seja possível calcular
                tipo_sql = tipo_sql.format(max_length)
            colunas.append(f"[{coluna}] {tipo_sql}")
        else:
            colunas.append(f"[{coluna}] VARCHAR(MAX)")  # Tipo padrão para casos não mapeados
    
    sql_create = f"CREATE TABLE {tabela} (\n" + ",\n".join(colunas) + "\n);"
    return sql_create

# Gerando o script SQL
script_sql_create = gerar_sql_create_table(dadosProvasObjLimpos, tabela)

# Exibe o script SQL, basta apenas copiar o result e colar no SSMS que será criado a tabela
print(script_sql_create)
```
